```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
       
    m_curr = b_curr = 0                      # star with random vaue of m and b
    iterations = 10000                         # initial iteration , try and error
    n = len(x)
    learning_rate = 0.01
    
    
    for i in range(iterations):
        # y prediction (y=mx+b)
        yp = m_curr*x + b_curr

        # cost function or mse
        cost = (1/n) * sum([val**2 for val in (y-yp)])
        
        # Partial derivative of m and b 
        md = -(2/n) * sum(x*(y-yp))
        bd = -(2/n) * sum(y-yp)
        
        m_curr = m_curr - learning_rate * md    # equation for step in curve
        b_curr = b_curr - learning_rate * bd
        
        plt.plot(x,yp)
        #print("m{}, b{}, iterations{}".format(m_curr, b_curr, iterations))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])

gradient_descent(x,y)
```
